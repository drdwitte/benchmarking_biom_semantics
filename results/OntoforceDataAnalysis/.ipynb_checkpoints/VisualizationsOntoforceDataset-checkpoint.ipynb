## 1. Instances per class

```python
df_class = pd.read_csv('OntoforceInstancesPerClass.txt', sep="\t", header=None)
```

```python
df_class.columns = ["class", "NumberOfInstancesPerClass"]

def powerOfTen(x):
    return math.ceil(math.log10(x))

df_class['InstancesPowerCount'] = df_class['NumberOfInstancesPerClass'].apply(lambda l: powerOfTen(l))
```

## 2. Property distribution per prefix

```python
import re
prefix_dictionary = {}
with open('OntoforceProperties.txt') as f:
    for line in f:
        m = re.search('^http:\/\/([a-zA-Z0-9\.-]+)\/', line)
        
        prefix = m.group(1)
        
        if prefix not in prefix_dictionary:
            prefix_dictionary[prefix] = 1
        else:
            prefix_dictionary[prefix] = prefix_dictionary[prefix] + 1

ser = pd.Series(prefix_dictionary)
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,10))

ser.plot(kind='bar', figsize=(20,10), ax=ax, fontsize=16, logy=True, title="Number of properties per prefix" )

fig.savefig('PropertiesPerPrefix.svg', format='svg', bbox_inches='tight', dpi=1200)
fig.savefig('PropertiesPerPrefix.eps', format='eps', bbox_inches='tight', dpi=1200)
```

## 3. Triples per property

```python
df_triples = pd.read_csv('OntoforceTriplesPerProperty.txt', sep="\t", header=None)
```

```python
df_triples.columns = ["Property", "TriplesPerProperty"]
```

```python
df_triples['TriplesPowerCount'] = df_triples['TriplesPerProperty'].apply(lambda l: powerOfTen(l))
```

```python
fig, ax = plt.subplots(1,2, figsize=(20,10))

ticks1 = [">=1", ">=10", ">=1e2",">=1e3",">=1e4",">=1e5",">=1e6",">=1e7"]

dfhandle = df_class[['InstancesPowerCount']].plot(kind='hist', bins=8, logy=False, ax=ax[0], legend=False, \
                                                  title="Distribution of the number of instances per class", \
                                                 ylim=[1,1300])
#dfhandle.set_xlabel(label)
ax[0].set_xticklabels(ticks1)
ax[0].set_xticks(np.arange(0,8) + 0.5)

dfhandle2 = df_triples[['TriplesPowerCount']].plot(kind='hist', bins=9, logy=False, ax=ax[1], legend=False, \
                                                   title="Distribution of the number of triples per predicate", \
                                                  ylim=[0,370])
#dfhandle2.set_xlabel(label2)

ticks2 = [">=1", ">=10", ">=1e2",">=1e3",">=1e4",">=1e5",">=1e6",">=1e7",">=1e8"]
ax[1].set_xticklabels(ticks2)
ax[1].set_xticks(np.arange(0,9) + 0.5)


fig.savefig('TripleDistributions.svg', format='svg', bbox_inches='tight', dpi=1200)
fig.savefig('TripleDistributions.eps', format='eps', bbox_inches='tight', dpi=1200)
```

## 4. Triples per prefix

```python
def extractPrefix(prop):
    m = re.search('^http:\/\/([a-zA-Z0-9\.-]+)\/', prop)
    return m.group(1)
```

```python
df_triples['prefix'] = df_triples['Property'].map(extractPrefix)
```

```python
df_prefix_triple_counts = df_triples.groupby('prefix', as_index=False).sum()
```

```python
df_prefix_triple_counts.index = df_prefix_triple_counts['prefix'].values
```

```python
df_prefix_triple_counts[['TriplesPerProperty']].plot(kind='barh', xlim=[1,1.1e9], logx=True, figsize=(5,8), fontsize=12, legend=False, title='#Triples per prefix')
```
